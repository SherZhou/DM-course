```python
import numpy as np
from sklearn.datasets import load_iris
# load data
iris=load_iris()
iris.data
```

```python
features=iris.data.T
# mean
miu=np.array([np.mean(features[0]),np.mean(features[1]),np.mean(features[2]),np.mean(features[3])])
miu
```

```python
# K1  homogeneous quadratic kernel matrix
K1=np.zeros((150,150))
for i in range(len(iris.data)):
    for j in range(len(iris.data)):
        K1[i][j]=np.dot(iris.data[i],iris.data[j])
K1
```

```python
# K2 centered kernel matrix
m1=np.identity(150)
m2=np.ones((150,150))/150
m3=m1-m2
K2=np.dot(np.dot(m3,K1),m3)
K2
```

```python
# K3 normalized kernel matrix
m4=np.eye(150,k=0)
for i in range(150):
    m4[i][i]=K2[i][i]**(-1/2)
K3=np.dot(np.dot(m4,K2),m4)
K3
```

```python
# Verify
#中心化
m5=iris.data
for i in range(150):
    m5[i]=iris.data[i]-miu    
m6=np.zeros((150,150))
for i in range(150):
    for j in range(150):
        m6[i][j]=np.dot(m5[i],m5[j])
m6
```

```python
# 归一化
m7=m5
for i in range(150):
    m7[i]=m5[i]/np.linalg.norm(m5[i])
m8=np.zeros((150,150))
for i in range(150):
    for j in range(150):
        m8[i][j]=np.dot(m7[i],m7[j])
m8
```
