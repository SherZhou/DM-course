```python
import numpy as np
from sklearn.datasets import load_iris
# load data
iris=load_iris()
features=iris.data.T
iris.data
```

```python
# K1  homogeneous quadratic kernel matrix
K1=np.zeros((150,150))
for i in range(len(iris.data)):
    for j in range(len(iris.data)):
        K1[i][j]=np.square(np.dot(iris.data[i],iris.data[j]))
K1
```

```python
# K2 centered kernel matrix
m1=np.identity(150)
m2=np.ones((150,150))/150
m3=m1-m2
K2=np.dot(np.dot(m3,K1),m3)
K2
```

```python
# K3 normalized kernel matrix
m4=np.identity(150)
for i in range(150):
    m4[i][i]=K1[i][i]**(-1/2)
K3=np.dot(np.dot(m4,K1),m4)
K3
```

```python
# Verify
# f将每个点转换到特征空间，再点积到的高维特征空间得f2
f=np.zeros((150,10))
for i in range(len(iris.data)):
    for j in range(0,4):
        f[i][j]=iris.data[i,j]**2
        for m in range(0,3):
            for n in range(m+1,4):
                j=j+1
                f[i,j]=2**0.5*iris.data[i,m]*iris.data[i,n]
                
f2=np.zeros((150,150))
for i in range (len(iris.data)):
    for j in range (len(iris.data)):
        f2[i,j]=np.dot(f[i],f[j])
f2
```

```python
# 中心化
f3=np.zeros((150,10))
f4=np.zeros((150,150))
miu=np.array([np.mean(f.T[0]),np.mean(f.T[1]),np.mean(f.T[2]),np.mean(f.T[3]),np.mean(f.T[4]),np.mean(f.T[5]),np.mean(f.T[6]),np.mean(f.T[7]),np.mean(f.T[8]),np.mean(f.T[9])])
for i in range(150):
    f3[i]=f[i]-miu 
for i in range (150):
    for j in range (150):
        f4[i,j]=np.dot(f3[i],f3[j])
f4
```

```python
# 归一化
f5=np.zeros((150,10))
f6=np.zeros((150,150))
for i in range(150):
    f5[i]=f[i]/np.linalg.norm(f[i])
for i in range(150):
    for j in range(150):
        f6[i][j]=np.dot(f5[i],f5[j])
f6
```
